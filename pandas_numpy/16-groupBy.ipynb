```python
import pandas as pd
```

```python
df_books = pd.read_csv('./data/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=',',header=0)
df_books.head(2)
```

### Similar al groupBy de SQL

```python
df_books.groupby('Author').count()  # agrupamos y contamos cuantos registro de cada author
```

```python
# .sum()  .mean() .min() .max() etc hace sobre las columnas de los registro agrupados
df_books.groupby('Author').max()
```

```python
# aqui nos muestra de cada author, los valores maximos de cada columna, si tienen varios registros, toma el mayor de entre ellos
```

#### Al hacer groupBy, el autor pasa ha ser el indice de cada row. por lo tanto se puede hacer loc de el nombre de dicho autor

```python
df_books.groupby('Author').sum().loc['William Davis']  # acumula los valores para el autor
```

```python
# se puede usar el metodo reset_index() para que genere indices numericos standard y autor pase a ser una columna
df_books.groupby('Author').max().reset_index()
```

## groupBy().agg()
    - En agg se le especifican las funciones de agregacion que quieras
    - ['min','max']

```python
# genera una columna  por cada funcion de agregacion que le indiquemos asociada a cada columna
# en este caso min y max
df_books.groupby('Author').agg(['min','max'])
```

### agg()
    - Acepta en forma de diccionario los nombre de las columnas como key y las funciones de agregacion como value
    - Se puede especificar cualquer numero de keys(columnas) , values(funciones)

```python
df_books.groupby('Author').agg({'Reviews':['mean','max'], 'User Rating':'max'})
```
