## Apply:
- Aplica funciones a nuestro DataFrame, funciones definiadas en python
    - Es muchisimo mas rapido que iterar por el DataFrame 

```python
import pandas as pd
```

```python
df_books = pd.read_csv('./data/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=',',header=0)
df_books.head(2)
```

```python
def two_times(value):
    return value * 2
```

```python
df_books['User Rating'].apply(two_times) # le apliamos la funcion a cada uno de los valores de la columna user rating
```

```python
# la salida se puede asignar a una nueva columna. Y la funcion puede evaluar y devolve lo que quieras

# Se puede usar lambda
df_books['Rating_2'] = df_books['User Rating'].apply(lambda x: x*3)
```

```python
df_books.head()
```

```python
# aplicamos a todas las columnas
df_books['Rating_2'] = df_books.apply(lambda x : x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis= 1)
 # hay que indicarle que le estamos haciendo referencia a las columnas con axis = 1
```
