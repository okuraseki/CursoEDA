### drop_duplicates()
    - Elimina los duplicados

```python
#pd.concat([df_books,df_books])  # estamos agregando el registro [0] al dataFrame
# ahora tendremos ese registro duplicado
df_books = pd.concat([df_books,df_books.loc[0:0]])
```

```python
df_books  # aqui tenemos el primer registro duplicado
```

```python
df_books.drop_duplicates()  # keep='last'  mantiene el ultimo encontrado y borra los anteriores duplicados
```
