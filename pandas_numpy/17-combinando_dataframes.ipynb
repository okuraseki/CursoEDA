```python
import pandas as pd
import numpy as np
```

## Metodos para unificar DataFrames
- Merge - Join:
    - LeftJoin : Trae todos los datos del DT de la izquierda incluyendo los datos en comun que tengan los 2
    - RigthJoin : Lo mismo pero le da la prioridad al DT del de la derecha
    - InnerJoin : Solo los datos que se encuentren en ambos DT, la interseccion
    - OuterJoin : Todos los datos, sin duplicar los datos que este repetidos

- Concat :
    - Axis 0 : Suma los row de los DT, siempre que las columnas coincidan
    - Axis 1 : Suma las columnas, los valores que no tengan se asignaran NaN (no null)

```python
dicc = {'A': ['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
  'C': ['C0','C1','C2','C3'],
   'D': ['D0','D1','D2','D3'] }
```

```python
dt1 = pd.DataFrame(dicc)
dt1
```

```python
dicc2 = {'A': ['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7'],
  'C': ['C4','C5','C6','C7'],
   'D': ['D4','D5','D6','D7'] }
```

```python
dt2 = pd.DataFrame(dicc2)
dt2
```

```python
# AÃ±ade las row del dt2 al dt1
pd.concat([dt1,dt2])
```

##### Los indices heredan el indice anterior, si queremos corregir esto... ignore_index=True

```python
pd.concat([dt1,dt2], ignore_index=True)
```

#### Tambien se puede cambiar el axis a 1 para unir por columnas


```python
pd.concat([dt1,dt2], axis=1)
```

## Merge


```python
dicc = {'key':['k0','k1','k2','k3'],
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3']}
dt_izqu = pd.DataFrame(dicc)
dt_izqu
```

```python
dicc = {'key':['k0','k1','k2','k3'],
       'C': ['C0','C1','C2','C3'],
   'D': ['D0','D1','D2','D3']}
dt_dere = pd.DataFrame(dicc)
dt_dere
```

```python
# combinando con merge
dt_izqu.merge(dt_dere) 
# por defecto usa la columna que tenga en comun 
```

```python
dicc = {'key':['k0','k1','k2','k3'],
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3']}
dt_izqu = pd.DataFrame(dicc)

dicc = {'key2':['k0','k1','k2','k3'],
       'C': ['C0','C1','C2','C3'],
   'D': ['D0','D1','D2','D3']}
dt_dere = pd.DataFrame(dicc)
```

```python
# Ahora la columna key no esta en los 2, por lo tanto hay que especificar como hacer el merge, o nos darÃ¡ error

dt_izqu.merge(dt_dere, left_on='key', right_on='key2') # especificamos que columna debe hacer match en el dt izq y en el dt derecho con letf_on y rigth_on
```

```python
dicc = {'key':['k0','k1','k2','k3'],
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3']}
dt_izqu = pd.DataFrame(dicc)

dicc = {'key2':['k0','k1','k2',np.nan],
       'C': ['C0','C1','C2','C3'],
   'D': ['D0','D1','D2','D3']}
dt_dere = pd.DataFrame(dicc)
```

```python
dt_izqu.merge(dt_dere, left_on='key', right_on='key2')
# Al tener el dt_derecho un key nan no coje ese row para ninguna de los 2, puesto que por defecto how es inner
```

```python
# how = 'left' / 'inner' / 'right'
dt_izqu.merge(dt_dere, left_on='key', right_on='key2', how='left')
# al especificar left, coje los valores de la izquierda para el row 3, pero en la derecha no
```
