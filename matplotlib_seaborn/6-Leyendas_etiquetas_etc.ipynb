```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(7,5)) # Define el tama√±o de figure
axes[0].plot(x,y, label = 'sin(x)') # defino label , pero hay que llamar al metodo legnend
axes[0].set_title('Relacion X - Y') # Define un titulo para el grafico
axes[0].set_xlabel('X') # Define label en el eje X
axes[0].set_ylabel('Y') # Define label en el eje y
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
```

```python
## Metodo sin OOP
plt.plot(x,y, label='sin(x)')
plt.title('Este es el titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right') # loc ubica la leyenda 
```
