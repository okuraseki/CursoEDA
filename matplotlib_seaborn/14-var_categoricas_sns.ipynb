```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(data=tips, x= 'day', hue='sex') # countplot sirve para contar cuantos de cada variable categorica
plt.show()
```

```python
sns.stripplot(data=tips, x= 'day',y='total_bill' ,hue='sex', dodge=True) # grafico de dispersion para variables categoricas
plt.show()
```

```python
sns.swarmplot(data=tips, x= 'day',y='total_bill' ,hue='sex', dodge=True) # parecido al anterior pero se abre cuando los datos esten concentrdados
```

```python
sns.boxplot(data=tips, x= 'day',y='total_bill' ,hue='sex', dodge=True) # como el boxplot de matplotlib
plt.show()
```

```python
sns.swarmplot(data=tips, x= 'day',y='total_bill' ,hue='sex', dodge=True) # grafico de dispersion para variables categoricas
sns.boxplot(data=tips, x= 'day',y='total_bill' ,hue='sex', dodge=True) # como el boxplot de matplotlib
plt.show()
# combinando 2 tipos de graficos
```

```python
sns.violinplot(data=tips, x= 'day',y='total_bill' ,hue='sex',split=True ,dodge=True) 
plt.show()
```

```python
sns.catplot(data=tips, x= 'day',y='total_bill' ,hue='sex',kind='box', col='time') # catplot es general, se le puede definir el tipo de grafico
plt.show()              # col nos permite separar en diferente graficos dependiendo de la column
```
