```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips') # datasets de prueba para aprender con seaborn
tips
```

```python
# data = el dataframe
sns.displot(data=tips, x= 'total_bill')
plt.show()
```

```python
# data = el dataframe
sns.displot(data=tips, x= 'total_bill', y= 'tip')
plt.show()
```

```python
# data = el dataframe
sns.displot(data=tips, x= 'total_bill', y= 'tip', hue='sex') # hue separa por colores en el column que le indiques
plt.show()
```

```python
sns.displot(data=tips, x= 'total_bill', hue='sex') # hue separa por colores en el column que le indiques
plt.show()
```

```python
sns.displot(data=tips, x= 'total_bill', hue='sex', kind='kde', legend=True, palette='dark', alpha=0.60) # kind tipo de histograma
plt.show()
```
