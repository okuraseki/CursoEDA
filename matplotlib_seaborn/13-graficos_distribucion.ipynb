```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips, x= 'tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge') # numero de barras, cumunlative va acumulando valores
plt.show()  # stat = count , probability , frequency, percent, density      # multiple = stack, layer, fill, dodge
```

```python
# diagrama de densidad
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True) #shade is deprecated use fill = True  
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count') # grafico escalonado
```

```python
# displot le indicamos kind = 'hist' , 'kde' etc...
sns.displot(data=tips, x='tip', hue='sex', kind='hist')  
plt.show()
```
