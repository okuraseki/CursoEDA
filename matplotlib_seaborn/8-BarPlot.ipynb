### El grafico de barra nos permite graficar variables categ√≥ricas
    - variables categoricas son de tipo texto

    - bar(varx, vary)
    - barh(varx, vary)  grafico de barras horizontal

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA','JAPON','MEXICO','COLOMBIA','ALEMANIA']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population, width=0.5, color=['red','blue']) # ancho de las barras
plt.show()  # color puede definir un color, si defines 2 va alternando, y puede definir 1 por cada variable categorica
```

```python
plt.bar(country,population, width=0.5, color=['red','blue','yellow','black','green']) 
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'), rotation = 45) # configuraciones del eje x
plt.show()  # rotation rota el nombre
```

```python
plt.barh(country,population)  # barh 
```
