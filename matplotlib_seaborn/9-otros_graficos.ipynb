```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

### Histograma
- Muestra la distribucion de los datos

```python
plt.hist(data, bins=50) # bins es le numero de barras
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step') # histtype tipo de histograma, bar , step 
plt.show()
```

### Boxplot

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True) # vertical , patch_artist rellena, notch enfoca los datos hace la mediana
plt.show()

# muestra rango de 25% en 25%, la linea de enmedio es la media
```

```python
data = np.append(data,200) # esto es un dato outliyer
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False) # si se pone a True, muestra el valoe 200
plt.show()
```

### Scatter o grafico de dispersion

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s = area, c = colors, marker='o', alpha=0.5
)  # s tamao, c color
plt.show()
```
