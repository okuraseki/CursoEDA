## Ventajas

- Velocidad.
- Poco codigo.
- Altamente customizable a nivel de graficas y visualizaciones.

### Estructura Basica

- sns.'Tipo de Grafica'(
		 data='Dataset',
		 x='Data en el eje x',
		 y='Data en el eje y',
		 hue='Variable de agrupamiento')

### Tipos de Graficas

- relplot (relacional): scatterplot, lineplot.
- displot (distribuciones): histplot, kdeplot, ecdfplot, rugplot.
- catplot (categorica): stripplot, swamplot, boxplot, violinplot, pointplot, barplot


## SET


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'], y=[1,3,2]) # seaborn esta montado sobre pyplot
plt.show()
```

```python
sns.set(style='dark', palette='dark', font_scale=1) # font para fuente
sns.set_theme(style='dark',palette='muted') # set esta deprecado , usar set_theme
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```
