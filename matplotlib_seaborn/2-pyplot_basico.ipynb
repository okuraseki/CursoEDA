## Pyplot
### Es una herramienta que tiene matplotlib, para dise√±ar y ejecutar graficas de manera sencilla.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)

y = x**2
```

- plt.plot(x, y)  Define la x y la y a la hora de dibujar la grafica, lo que tomara como x e y
- plt.show() Dibuja la grafica

```python
plt.plot(x,y)
plt.show() 
```

```python
plt.plot(x,y, 'ro-') 
# Definimos el color con la primera r rojo, b blue, etc..
# el segundo caracter s square , 0 cirulo, x dibuja x
# - une los puntos
# : linea punteada
plt.show() 
```

```python
plt.hist(x)
plt.show() 
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
