```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
for style in plt.style.available:
    print(style)

    # Estilos disponibles
```

```python
plt.style.use('ggplot')
fig , ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r')
ax.plot(x,x+2,'b')
ax.plot(x,x+3,'y')
ax.plot(x,x+4,'g')

```

### Modificaciones tipo Matlab

```python
plt.style.use('dark_background')
fig , ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'g:')
```

### Con pyplot

```python
fig , ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color= 'green') # acepta codigo rgb hex 
ax.plot(x,x+2,color= 'blue', alpha=0.2) # alpha define la transparencia entre 0 y 1
ax.plot(x,x+3,color= 'yellow', linewidth=5) #grosor de la linea
ax.plot(x,x+4,color= 'red', linestyle='dashed') # el style de la linea
ax.plot(x,x+5,color= 'white', linestyle=':', marker='P', markersize=8, markerfacecolor='red') # el tipo de punto, tamaa del marker, color del marker
```
