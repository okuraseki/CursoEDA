```python
import pandas as pd 
import altair as alt 
```

```python
chicago_data = './data/Traffic_Crashes1.csv'
df_chicago = pd.read_csv(chicago_data)
```

```python
df_chicago['CRASH_DATE_time'] = df_chicago['CRASH_DATE'].apply(lambda x : pd.to_datetime(x, errors='coerce', utc=True))
```

```python
df_chicago.head()
```

```python
# Uso de groupBy

report_1 = df_chicago.groupby(['LIGHTING_CONDITION','REPORT_TYPE','CRASH_HOUR']).agg({'NUM_UNITS':'sum'})
```

```python
report_1
```

```python
report_1 = report_1.reset_index() # para generar los indices
```

```python
report_1
```

```python
# Graficos
alt.Chart(report_1).mark_bar().encode(x='LIGHTING_CONDITION',y='NUM_UNITS',color='REPORT_TYPE').properties(width=220)
```

```python
# Varias funciones de agregacion sobre un mismo groupby
report_1 = df_chicago.groupby(['LIGHTING_CONDITION','REPORT_TYPE','CRASH_HOUR']).agg({'NUM_UNITS':['sum','min','max']})
```
